```python
def plot(n_result, y_result, s_result, J_end, I_end, t_add): 
    import warnings
    warnings.filterwarnings("ignore")
    from matplotlib.axes._axes import _log as matplotlib_axes_logger
    matplotlib_axes_logger.setLevel('ERROR')

    import matplotlib.pyplot as plt
    import numpy as np
    from matplotlib import cm
    from matplotlib.patches import Circle
    from matplotlib.offsetbox import (TextArea, DrawingArea, OffsetImage,
                                        AnnotationBbox)
    from matplotlib.cbook import get_sample_data

    import numpy as np



    s_j = np.array([J_end-1.4]*(2))
    I = range(I_end)



    fig,axes = plt.subplots(nrows=14, ncols=1, figsize=(10, 14*(I_end-1)), sharex=True, sharey=True)
    # ax=fig.add_axes([0,0,5,0.5])


    for tt, ax in enumerate(axes.flatten()):
        tt=tt +t_add
        # ax.set(xlim=[-1,J_end+1],ylim=[-0.5,3.5])
        # plt.figure(figsize=(12, I_end-1))

        if s_result['s'][s_result['t'] == tt].tolist()[0] == 0:
            color_s = 'red'
            mark = 'x'
            mark_1 = 'x'
        elif s_result['s'][s_result['t'] == tt].tolist()[0] == 1:
            color_s = 'green'
            mark = 9
            mark_1 = 0
        elif s_result['s'][s_result['t'] == tt].tolist()[0] == 2:
            color_s = 'green'
            mark = 10
            mark_1 = 3

        i_d_1 = n_result['i'][(n_result['t'] == tt) & (n_result['d'] == 1) & (n_result['value'] > 0)].tolist()
        i_d__1 = n_result['i'][(n_result['t'] == tt) & (n_result['d'] == -1) & (n_result['value'] > 0)].tolist()
        
        j_d_1 = n_result['j'][(n_result['t'] == tt) & (n_result['d'] == 1) & (n_result['value'] > 0)].tolist()
        j_d__1 = n_result['j'][(n_result['t'] == tt) & (n_result['d'] == -1) & (n_result['value'] > 0)].tolist()
    

        plt.ylim(-0.5, I_end-0.5)
        plt.xlim(-0.5,J_end+2.5)

        for i in I:
            ax.plot([-0.5,J_end+0.5],[i+0.5,i+0.5],c='k',linestyle='--')

        for ind in range(len(i_d__1)):
            j = j_d__1[ind]
            i = i_d__1[ind]
            n_r_n = n_result['value'][(n_result['t'] == tt) & (n_result['i'] == i) & (n_result['j'] ==j) & (n_result['d']==-1)].tolist()[0]
            map_vir_r_n = cm.get_cmap(name='Reds')
            color_r_n = map_vir_r_n(n_r_n)

        
            ax.scatter(j,i,marker = 's',s=500, c = 'salmon',zorder=1, hatch =round(n_r_n/5+0.5)*'-',edgecolor='k',alpha = 0.5)

        for ind in range(len(i_d_1)):
            j = j_d_1[ind]
            i = i_d_1[ind]   
            n_b_n = n_result['value'][(n_result['t'] == tt) & (n_result['i'] == i) & (n_result['j'] ==j) & (n_result['d']==1)].tolist()[0] 
            map_vir_b_n = cm.get_cmap(name='Blues')
            color_b_n = map_vir_b_n(n_b_n)

            ax.scatter(j,i,marker = 's',s=500,c = 'skyblue',zorder=1, hatch = round(n_b_n/5+0.5)*'/',edgecolor='k',alpha = 0.5 )
    

        y_i_1 = y_result['i'][(y_result['t'] == tt) & (y_result['d'] == 1) & (y_result['value'] > 0)].tolist()
        y_j_1 = y_result['j'][(y_result['t'] == tt) & (y_result['d'] == 1) & (y_result['value'] > 0)].tolist()
        y_id_1 = y_result['i_d'][(y_result['t'] == tt) & (y_result['d'] == 1) & (y_result['value'] > 0)].tolist()
        y_jd_1 = y_result['j_d'][(y_result['t'] == tt) & (y_result['d'] == 1) & (y_result['value'] > 0)].tolist()
        for ind in range(len(y_j_1)):
            yn_b = y_result['value'][(y_result['t'] == tt) & (y_result['d'] == 1) & (y_result['value'] > 0) & (y_result['i'] == y_i_1[ind]) & (y_result['j'] == y_j_1[ind]) & (y_result['j_d'] == y_jd_1[ind]) & (y_result['i_d'] == y_id_1[ind])].tolist()[0]
            map_vir_b = cm.get_cmap(name='Blues')
            color_b_y = map_vir_b(yn_b)
            ax.plot([y_j_1[ind]+0.25,y_jd_1[ind]-0.25],[y_i_1[ind],y_id_1[ind]],c='skyblue',zorder=0,lw = yn_b*1.5+0.5,alpha = 0.5)

        y_i__1 = y_result['i'][(y_result['t'] == tt) & (y_result['d'] == -1) & (y_result['value'] > 0)].tolist()
        y_j__1 = y_result['j'][(y_result['t'] == tt) & (y_result['d'] == -1) & (y_result['value'] > 0)].tolist()
        y_id__1 = y_result['i_d'][(y_result['t'] == tt) & (y_result['d'] == -1) & (y_result['value'] > 0)].tolist()
        y_jd__1 = y_result['j_d'][(y_result['t'] == tt) & (y_result['d'] == -1) & (y_result['value'] > 0)].tolist()
        for ind in range(len(y_j__1)):
            yn_r = y_result['value'][(y_result['t'] == tt) & (y_result['d'] == -1) & (y_result['value'] > 0) & (y_result['i'] == y_i__1[ind]) & (y_result['j'] == y_j__1[ind]) & (y_result['j_d'] == y_jd__1[ind]) & (y_result['i_d'] == y_id__1[ind])].tolist()[0]
            map_vir_r = cm.get_cmap(name='Reds')
            color_r_y = map_vir_r(yn_r)
            ax.plot([y_j__1[ind]+0.25,y_jd__1[ind]-0.25],[y_i__1[ind],y_id__1[ind]],c='salmon',zorder=0, lw = yn_r*1.5+0.5,alpha = 0.5)

        ax.plot(s_j,[-0.5,3.6],c = color_s,ls='-',lw=5)

        ax.plot(np.array([J_end+2.5]*(2)),[-0.5,1.5],ls='-',lw=160,c = 'grey',zorder=0)
        ax.scatter([J_end+2,J_end+2,J_end+2,J_end+2],[0,1,2,3],s = 500,marker = mark,c = 'w',zorder=10)
        ax.scatter([J_end+2,J_end+2,J_end+2,J_end+2],[0,1,2,3],s = 500,marker = mark_1,c = 'w',zorder=10)

        if mark == 'x':
            ax.scatter([J_end+2,J_end+2,J_end+2,J_end+2],[0,1,2,3],s = 1000,marker = 'o',c = 'w',zorder=8)
            ax.scatter([J_end+2,J_end+2,J_end+2,J_end+2],[0,1,2,3],s = 800,marker = 'o',c = 'grey',zorder=9)
        

        ax.set_title('time '+str(tt),fontsize=20)
    fig.savefig('sorting3_1_color.jpg',bbox_inches="tight", pad_inches=0.0)
    
    return
```
